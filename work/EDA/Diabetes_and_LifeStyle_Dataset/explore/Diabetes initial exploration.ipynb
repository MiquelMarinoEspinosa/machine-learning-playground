```python
import pandas as pd
from sklearn.model_selection import train_test_split

diabetes_data_set = pd.read_csv('../../../datasets/Diabetes_and_LifeStyle_Dataset.csv')
diabetes_data_set
```

```python
y = diabetes_data_set['diagnosed_diabetes']
X = diabetes_data_set.copy()
X.drop(columns = ['diagnosed_diabetes'], inplace=True)
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.4,
    random_state=42
)
X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)
diabetes_data_set_train = pd.concat([X_train, y_train], axis=1)
diabetes_data_set_train
```
