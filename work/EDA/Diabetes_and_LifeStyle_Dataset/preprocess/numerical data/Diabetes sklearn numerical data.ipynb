```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler

diabetes_dataset = pd.read_csv('../../../../datasets/Diabetes_and_LifeStyle_Dataset.csv')
diabetes_dataset
```

```python
y = diabetes_dataset['diagnosed_diabetes']
X = diabetes_dataset.copy()
X.drop(columns=['diabetes_stage','diagnosed_diabetes'], inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)


print(
    X_train.shape,
    y_train.shape,
    X_test.shape,
    y_test.shape,
    X_val.shape,
    y_val.shape
)
```

```python
num_cols_high_variance = [col for col in X_train.columns 
                          if X_train[col].dtype in ['int64', 'float64']
                          and X_train[col].std() > 1]
X_train[num_cols_high_variance].std(numeric_only=True)
```

```python
X_train_min_max_scaled = X_train.copy()

min_max_scaler = MinMaxScaler()

X_train_min_max_scaled[num_cols_high_variance] = min_max_scaler.fit_transform(X_train_min_max_scaled[num_cols_high_variance])

X_train_min_max_scaled.std(numeric_only=True)
```
