```python
import pandas as pd
from sklearn.model_selection import train_test_split

diabetes_data_set = pd.read_csv('../../../../datasets/Diabetes_and_LifeStyle_Dataset.csv')
diabetes_data_set
```

```python
X = diabetes_data_set
y = diabetes_data_set['diagnosed_diabetes']
X.drop(['diabetes_stage', 'diagnosed_diabetes'], axis=1, inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)

print(
    X_train.shape, 
    y_train.shape, 
    X_test.shape,
    y_test.shape,
    X_val.shape,
    y_val.shape
)
```

```python
num_cols = [col for col in X_train.columns
            if X_train[col].dtype in ['int64', 'float64']
]
print(num_cols, len(num_cols), sep='\n\n')
```

```python
num_cols_high_variance = [col for col in num_cols
                         if X_train[col].std() > 1]
print(num_cols_high_variance, len(num_cols_high_variance), sep='\n\n')
```

```python
#min max scale
X_train_min_max_scale = X_train.copy()
for col in num_cols_high_variance:
    X_train_min_max_scale[col] = (X_train_min_max_scale[col] - X_train_min_max_scale[col].min()) / (X_train_min_max_scale[col].max() - X_train_min_max_scale[col].min())
X_train_min_max_scale.std(numeric_only=True)
```

```python
X_train_z_score_scale = X_train.copy()

for col in num_cols_high_variance:
    X_train_z_score_scale[col] = (X_train_z_score_scale[col] - X_train_z_score_scale[col].mean()) / X_train_z_score_scale[col].std()
X_train_z_score_scale.std(numeric_only=True)
```
