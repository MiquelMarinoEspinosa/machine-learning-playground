```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np

bmw_data_set = pd.read_csv('../../../datasets/BMW sales data (2010-2024).csv')
bmw_data_set
```

```python
X = bmw_data_set
y = X.Price_USD
X.drop(columns=['Price_USD'], inplace=True)
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y,
    train_size=0.60,
    random_state=42
)
X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)
bmw_data_set_train = pd.concat([X_train, y_train], axis=1)
bmw_data_set_train
```

```python
pd.options.display.max_rows = 20
bmw_data_year_model = bmw_data_set_train.groupby(by=['Year', 'Model'])
bmw_data_year_model.Price_USD.agg(['mean', 'max', 'min'])
```

```python
bmw_data_year_model_price_mean = bmw_data_year_model.Price_USD.mean()
bmw_data_year_model_price_mean_df = bmw_data_year_model_price_mean.to_frame()
bmw_data_year_model_price_mean_df = bmw_data_year_model_price_mean_df.reset_index()
bmw_data_year_model_price_mean_df
```

```python
bmw_data_year_model_price_mean_df.columns
```

```python
max_price = np.argmax(bmw_data_year_model_price_mean_df['Price_USD'])
bmw_data_year_model_price_mean_df.loc[max_price]
```

```python
min_price = np.argmin(bmw_data_year_model_price_mean_df['Price_USD'])
bmw_data_year_model_price_mean_df.loc[min_price]
```

```python
models_expensive = []
for year in bmw_data_year_model_price_mean_df.Year.unique():
    bmw_data_year_df = bmw_data_year_model_price_mean_df[bmw_data_year_model_price_mean_df['Year'] == year]
    bmw_data_year_max_idx = np.argmax(bmw_data_year_df.Price_USD)
    print(f"{bmw_data_year_df.iloc[bmw_data_year_max_idx]}")
    models_expensive.append(bmw_data_year_df.iloc[bmw_data_year_max_idx]['Model'])
    print("")
print(f"Most expensive models: {models_expensive}")
```

```python
models_cheapest = []
for year in bmw_data_year_model_price_mean_df.Year.unique():
    bmw_data_year_df = bmw_data_year_model_price_mean_df[bmw_data_year_model_price_mean_df['Year'] == year]
    bmw_data_year_min_idx = np.argmin(bmw_data_year_df.Price_USD)
    print(f"{bmw_data_year_df.iloc[bmw_data_year_min_idx]}")
    models_cheapest.append(bmw_data_year_df.iloc[bmw_data_year_min_idx]['Model'])
    print("")
print(f"Cheapest models: {models_cheapest}")
```

```python
print(f"Most expensive models: {models_expensive}")
print(f"Cheapest models: {models_cheapest}")
```

```python
models_expensive_and_cheapeast = np.intersect1d(models_expensive, models_cheapest)
models_expensive_and_cheapeast
```

```python
print("Models just expensive")
[model for model in models_expensive if model not in models_expensive_and_cheapeast]
```

```python
print("Models just cheap")
[model for model in models_cheapest if model not in models_expensive_and_cheapeast]
```

```python
print(f"Price max - min difference")
for year in bmw_data_year_model_price_mean_df.Year.unique():
    bmw_data_year_df = bmw_data_year_model_price_mean_df[bmw_data_year_model_price_mean_df['Year'] == year]
    bmw_data_year_max_idx = np.argmax(bmw_data_year_df.Price_USD)
    bmw_data_year_min_idx = np.argmin(bmw_data_year_df.Price_USD)
    price_diff = bmw_data_year_df.iloc[bmw_data_year_max_idx]['Price_USD'] - bmw_data_year_df.iloc[bmw_data_year_min_idx]['Price_USD'] 
    print(f"Year: {year} / Difference: {price_diff:.2f}")
    print("")
```
