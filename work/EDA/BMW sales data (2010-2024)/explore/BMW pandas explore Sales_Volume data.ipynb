```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

bmw_sales_data = pd.read_csv('../../../datasets/BMW sales data (2010-2024).csv')
bmw_sales_data
```

```python
X = bmw_sales_data.copy()
y = X['Sales_Volume']
X.drop(columns=['Sales_Volume'], inplace = True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)
bmw_sales_train_data_set = pd.concat([X_train, y_train], axis=1)
bmw_sales_train_data_set
```

```python
bmw_sales_volume_year_model = bmw_sales_train_data_set.groupby(by=['Year', 'Model']).Sales_Volume.mean().reset_index()
bmw_sales_volume_year_model
```

```python
models_more_sales = []
print("Max sales per year")
for year in bmw_sales_volume_year_model.Year.unique():
    bmw_sales_year = bmw_sales_volume_year_model['Year'] == year
    bmw_sales_year_model = bmw_sales_volume_year_model[bmw_sales_year]
    max_index = np.argmax(bmw_sales_year_model.Sales_Volume)
    bmw_sales_year_model_max = bmw_sales_year_model.iloc[max_index]
    print(f'Year: {year}')
    print(f'{bmw_sales_year_model_max}')
    models_more_sales.append(bmw_sales_year_model_max['Model'])
    print("")
print(f'{models_more_sales}')
```

```python
models_less_sales = []
print("Min sales per year")
for year in bmw_sales_volume_year_model.Year.unique():
    bmw_sales_year = bmw_sales_volume_year_model['Year'] == year
    bmw_sales_year_model = bmw_sales_volume_year_model[bmw_sales_year]
    min_index = np.argmin(bmw_sales_year_model.Sales_Volume)
    bmw_sales_year_model_min = bmw_sales_year_model.iloc[min_index]
    print(f'Year: {year}')
    print(f'{bmw_sales_year_model_min}')
    models_less_sales.append(bmw_sales_year_model_min['Model'])
    print("")
print(f'{models_less_sales}')
```

```python
print("Max sales models")
print(models_more_sales)
print("Less sales models")
print(models_less_sales)
```

```python
models_more_and_less = np.intersect1d(models_more_sales, models_less_sales)
```
