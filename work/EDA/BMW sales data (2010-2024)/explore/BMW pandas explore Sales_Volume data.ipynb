```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

bmw_sales_data = pd.read_csv('../../../datasets/BMW sales data (2010-2024).csv')
bmw_sales_data
```

```python
X = bmw_sales_data.copy()
y = X['Sales_Volume']
X.drop(columns=['Sales_Volume'], inplace = True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)
bmw_train_data_set = pd.concat([X_train, y_train], axis=1)
bmw_train_data_set
```

```python
bmw_sales_volume_year_model = bmw_train_data_set.groupby(by=['Year', 'Model']).Sales_Volume.mean().reset_index()
bmw_sales_volume_year_model
```
