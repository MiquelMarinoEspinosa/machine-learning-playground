```python
import pandas as pd
#import matplotlib
import numpy as np
%matplotlib inline
# only in a Jupyter notebook
import matplotlib.pyplot as plt

df = pd.read_csv('../../../datasets/BMW sales data (2010-2024).csv')
df.info()
```

```python
plt.figure(figsize=(15, 5), dpi=80)
for model in np.sort(df.Model.unique()):
    plt.plot(df[df.Model == model].groupby(by='Year').Sales_Volume.mean())
plt.title('Sales Volume by model and year')
plt.legend(np.sort(df.Model.unique()), loc='lower left', fontsize='x-small')
plt.show()
```

```python
for model in np.sort(df.Model.unique()):
    plt.figure(figsize=(15, 5), dpi=80)
    plt.plot(df[df.Model == model].groupby(by='Year').Sales_Volume.mean())
    plt.legend([model], loc='lower right')
    plt.title('Sales Volume per ' + model + ' by year')
    plt.show()
```

```python
for model in np.sort(df.Model.unique()):
    plt.figure(figsize=(15, 5), dpi=80)
    plt.plot(df[df.Model == model].groupby(by='Year').Price_USD.mean())
    plt.legend([model], loc='lower right')
    plt.title('Avg Price per ' + model + ' by year')
    plt.show()
```
