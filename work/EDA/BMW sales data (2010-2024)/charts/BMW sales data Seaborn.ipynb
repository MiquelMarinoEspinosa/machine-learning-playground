```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../../../datasets/BMW sales data (2010-2024).csv', index_col='Year')
df.head()
```

```python
df_price_group_by_year_model = df.groupby(by=['Model','Year']).Price_USD.mean().reset_index(name='Price_Mean')
plt.figure(figsize=(15,8))
price_line_plot = sns.lineplot(
    data=df_price_group_by_year_model, 
    x='Year', 
    y='Price_Mean', 
    hue='Model'
)
price_line_plot.set(title='Line plot Price per year and model')
sns.move_legend(price_line_plot, 'upper left', bbox_to_anchor=(1, 1))
```

```python
df_price_group_by_year_model = df.groupby(by=['Model','Year']).Price_USD.mean().reset_index(name='Price_Mean')
plt.figure(figsize=(15,8))
price_line_plot = sns.barplot(
    data=df_price_group_by_year_model, 
    x='Year', 
    y='Price_Mean', 
    hue='Model'
)
price_line_plot.set(title='Line plot Price per year and model')
sns.move_legend(price_line_plot, 'upper left', bbox_to_anchor=(1, 1))
```

```python
df_sales_group_by_year_model = df.groupby(by=['Model','Year']).Sales_Volume.mean().reset_index(name='Sales_Volume')
plt.figure(figsize=(15,8))
sales_line_plot = sns.lineplot(
    data=df_sales_group_by_year_model, 
    x='Year', 
    y='Sales_Volume', 
    hue='Model'
)
sales_line_plot.set(title='Line Plot Sales per year and model')
sns.move_legend(sales_line_plot, 'upper left', bbox_to_anchor=(1, 1))
```

```python
df_sales_group_by_year_model = df.groupby(by=['Model','Year']).Sales_Volume.mean().reset_index(name='Sales_Volume')
plt.figure(figsize=(15,8))
sales_bar_plot = sns.barplot(
    data=df_sales_group_by_year_model, 
    x='Year', 
    y='Sales_Volume', 
    hue='Model'
)
sales_bar_plot.set(title='Bar plot Sales Volume per year and model')
sns.move_legend(sales_bar_plot, 'upper left', bbox_to_anchor=(1, 1))
```
