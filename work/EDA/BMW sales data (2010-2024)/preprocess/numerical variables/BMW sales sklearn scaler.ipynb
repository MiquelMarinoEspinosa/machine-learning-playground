```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import numpy as np

bmw_data_set = pd.read_csv('../../../../datasets/BMW sales data (2010-2024).csv')
bmw_data_set
```

```python
bmw_data_set.info()
```

```python
bmw_data_set.describe()
```

```python
min_max_scaler = MinMaxScaler()
mileage_km_min_max_scale = min_max_scaler.fit_transform(bmw_data_set['Mileage_KM'].to_numpy().reshape(-1, 1))
mileage_km_min_max_scale
```

```python
mileage_km_min_max_scale.shape
```

```python
mileage_km_min_max_scale_df = pd.DataFrame(mileage_km_min_max_scale, columns=['Mileage_KM_min_max_scale'])
mileage_km_min_max_scale_df
```

```python
bmw_data_set = pd.concat([bmw_data_set, mileage_km_min_max_scale_df], axis=1)
bmw_data_set
```

```python
bmw_data_set.Mileage_KM_min_max_scale.describe()
```

```python
standard_scaler = StandardScaler()
mileage_km_standard_scale = standard_scaler.fit_transform(bmw_data_set['Mileage_KM'].to_numpy().reshape(-1,1))
mileage_km_standard_scale
```

```python
mileage_km_standard_scale_df = pd.DataFrame(mileage_km_standard_scale, columns=['Mileage_KM_standard_scale'])
mileage_km_standard_scale_df
```

```python
bmw_data_set = pd.concat([bmw_data_set, mileage_km_standard_scale_df], axis=1)
bmw_data_set
```

```python
bmw_data_set.Mileage_KM_standard_scale.describe()
```
