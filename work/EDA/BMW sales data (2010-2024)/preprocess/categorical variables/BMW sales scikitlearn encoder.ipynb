```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
import numpy as np

bmw_data_set = pd.read_csv('../../../../datasets/BMW sales data (2010-2024).csv')
bmw_data_set.info()
```

```python
cat_cols_low_cardinality = [col for col in bmw_data_set.columns if bmw_data_set[col].dtype == 'object'
                          and len(bmw_data_set[col].unique()) <= 2]
cat_cols_low_cardinality
```

```python
ordinal_encoder = OrdinalEncoder(dtype=np.int64)
bmw_data_set[cat_cols_low_cardinality] = ordinal_encoder.fit_transform(bmw_data_set[cat_cols_low_cardinality])
bmw_data_set
```

```python
bmw_data_set.info()
```

```python
cat_cols_high_cardinality = [col for col in bmw_data_set.columns if bmw_data_set[col].dtype == 'object'
                            and len(bmw_data_set[col].unique()) > 2]
cat_cols_high_cardinality
```

```python
one_hot_encoder = OneHotEncoder(dtype=np.int64, sparse_output=False)
one_hot_encoded_cat_high_cardinality = pd.DataFrame(one_hot_encoder.fit_transform(bmw_data_set[cat_cols_high_cardinality]))
one_hot_encoded_cat_high_cardinality.columns = one_hot_encoder.get_feature_names_out()
one_hot_encoded_cat_high_cardinality
```

```python
data_set_num_cols = bmw_data_set.drop(cat_cols_high_cardinality, axis=1)
data_set_num_cols
```

```python
bmw_encoded_data_set = pd.concat([data_set_num_cols, one_hot_encoded_cat_high_cardinality], axis=1)
bmw_encoded_data_set
```
