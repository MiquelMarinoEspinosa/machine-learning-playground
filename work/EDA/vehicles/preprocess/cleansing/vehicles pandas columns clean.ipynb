```python
import pandas as pd
from sklearn.model_selection import train_test_split

vehicles_data_set = pd.read_csv('/home/jovyan/work/datasets/vehicles.csv')
vehicles_data_set.head()
```

```python
y = vehicles_data_set['price']
X = vehicles_data_set.copy()
X.drop(columns=[
    'price',
    'region_url', 
    'description', 
    'image_url', 
    'url',
    'lat',
    'long',
    'VIN',
    'posting_date',
    'county',
    #71.72% null values
    'size',
    #40.88% null values
    'condition',
    #41.64% null values
    'cylinders',
    #30.57% null values
    'paint_color',
    #30.6% null values
    'drive'
], inplace=True, errors='ignore')

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
cols_with_null = [col for col in X_train.columns
                if X_train[col].isnull().any()]

for col_with_null in cols_with_null:
    #print(X_train[col_with_null].value_counts(dropna=False))
    col_null_sum = X_train[col_with_null].isnull().sum()
    print(col_with_null, col_null_sum, round((col_null_sum / X_train.shape[0]) * 100, 2))
print(X_train['type'].value_counts(dropna=False))
```

```python
cat_cols = [col for col in X_train.columns if X_train[col].dtype == 'object']
for col in cat_cols:
    print(
        col,
        '--------',
        X_train[col].unique().size
    )
```
