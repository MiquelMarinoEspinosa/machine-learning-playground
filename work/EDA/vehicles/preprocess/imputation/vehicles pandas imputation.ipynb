```python
import pandas as pd
from sklearn.model_selection import train_test_split

vehicles_dataset = pd.read_csv('/home/jovyan/work/datasets/vehicles_cleaned.csv')
vehicles_dataset
```

```python
y = vehicles_dataset['price']
X = vehicles_dataset.copy()
X.drop(columns=['price'], inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
X_train.reset_index(inplace=True, drop=True)
cols_with_null = [col for col in X_train.columns
                 if X_train[col].isnull().any()]
cols_with_null, X_train.columns
```

```python
for col_with_null in cols_with_null:
    print(X_train[col_with_null].value_counts(dropna=False))
```

```python
#For the year let's input the value before or after
indexes_year_null_values = X_train[X_train['year'].isnull()].index.tolist()
for row_index in indexes_year_null_values:
    print(f'row_index: {row_index}')
    row_before = X_train.loc[(row_index - 1), 'year']
    print(f'year row before value: {row_before}')
    row_after = X_train.loc[(row_index + 1), 'year']
    print(f'year row before value: {row_after}')
```

```python
def print_years(df):
    years = []
    for row_index in indexes_year_null_values:
        years.append(df.loc[row_index, 'year'])
    print(years)
```

```python
X_train_imputate = X_train.copy()
X_train_imputate['year'].ffill(inplace=True)
print_years(X_train_imputate)
```

```python
X_train_imputate = X_train.copy()
X_train_imputate['year'].bfill(inplace=True)
print_years(X_train_imputate)
```
