```python
X = vehicles_data_set.copy()

y = vehicles_data_set['price'].copy()

X.drop(columns=['price'], inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.6,
    random_state=42
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
X_train.reset_index(inplace=True, drop=True)
```

```python
rows_with_model_null_value = X_train[X_train['manufacturer'].isnull()].index.to_list()

X_train_manufacturer_most_frequent_imputed = X_train.copy()

print(
    X_train_manufacturer_most_frequent_imputed['manufacturer'].isnull().any()
)

imputer_most_frequent = SimpleImputer(strategy='most_frequent', add_indicator=True)

imputer_most_frequent_values = imputer_most_frequent.fit_transform(
    np.reshape(
        X_train_manufacturer_most_frequent_imputed['manufacturer'],
        (-1,1)
    )
)

X_train_manufacturer_most_frequent_imputed.drop(
    columns=['manufacturer'],
    inplace=True
)

X_train_manufacturer_most_frequent_imputed = pd.concat(
    [
        X_train_manufacturer_most_frequent_imputed,
        pd.Series(
            imputer_most_frequent_values[:,0],
            name='manufacturer'
        ),
        pd.Series(
            imputer_most_frequent_values[:,1],
            name='manufacturer_missing'
        )
    ], axis=1
)

print(
    X_train_manufacturer_most_frequent_imputed['manufacturer'].isnull().any()
)

for row_with_model_null_value in rows_with_model_null_value:
    print(
        X_train['manufacturer'][row_with_model_null_value],
        '-------------',
        X_train_manufacturer_most_frequent_imputed['manufacturer'][row_with_model_null_value],
        '-------------',
        X_train_manufacturer_most_frequent_imputed['manufacturer_missing'][row_with_model_null_value],
    )
```
