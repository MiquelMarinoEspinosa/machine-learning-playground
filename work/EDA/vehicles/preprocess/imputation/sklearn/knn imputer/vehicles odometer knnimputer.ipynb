```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer

vehicles_data_set = pd.read_csv('/home/jovyan/work/datasets/vehicles_cleaned.csv')
vehicles_data_set.head()
```

```python
X = vehicles_data_set.copy()
y = vehicles_data_set['price'].copy()

X.drop(columns=['price'], inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
X_train.reset_index(inplace=True, drop=True)
```

```python
cols_with_null_values = [col for col in X_train.columns
                  if X_train[col].isnull().any()]

for col_with_null_values in cols_with_null_values:
    print(
        X_train[col_with_null_values].value_counts(dropna=False, normalize=True)
    )
```

```python
rows_with_odometer_with_null_values = X_train[X_train['odometer'].isnull()].index.to_list()

X_train_knn_imputed = X_train.copy()

print(
    X_train_knn_imputed['odometer'].isnull().any()
)

knn_imputer = KNNImputer()

knn_imputed_values = knn_imputer.fit_transform(
    np.reshape(
        X_train_knn_imputed['odometer'],
        (-1,1)
    )
)

X_train_knn_imputed.drop(columns=['odometer'], inplace=True)
```
