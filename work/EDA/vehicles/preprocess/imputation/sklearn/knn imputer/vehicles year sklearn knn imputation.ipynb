```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer
import numpy as np

vehicles_data_set = pd.read_csv('/home/jovyan/work/datasets/vehicles_cleaned.csv')
vehicles_data_set.head()
```

```python
X = vehicles_data_set.copy()
y = vehicles_data_set['price']
X.drop(columns=['price'], inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
cols_with_null = [col for col in X_train.columns
                 if X_train[col].isnull().any()]

for col_with_null in cols_with_null:
    print(X_train[col_with_null].value_counts(dropna=False))
```

```python
X_train.reset_index(inplace=True)
```

```python
rows_with_year_null = X_train[X_train['year'].isnull()].index.to_list()

X_train_year_knn_5_uniform_imputed = X_train.copy()

print(
    X_train_year_knn_5_uniform_imputed['year'].isnull().any()
)

knn_imputer_5_uniform = KNNImputer(copy=False, add_indicator=True)

knn_5_uniform_values = knn_imputer_5_uniform.fit_transform(
    np.reshape(X_train_year_knn_5_uniform_imputed['year'], (-1,1))
)

X_train_year_knn_5_uniform_imputed = pd.concat(
    [
        X_train_year_knn_5_uniform_imputed,
        pd.Series(
            knn_5_uniform_values[:,1],
            name='year_missing'
        )
    ], axis=1
)

print(
    X_train_year_knn_5_uniform_imputed['year'].isnull().any()
)

X_train_year_knn_5_uniform_imputed['year'] = X_train_year_knn_5_uniform_imputed['year'].astype(np.int64)

for row_with_year_null in rows_with_year_null:
    print(
        X_train_year_knn_5_uniform_imputed['year'][row_with_year_null],
        '------------',
        X_train_year_knn_5_uniform_imputed['year_missing'][row_with_year_null]
    )
```
