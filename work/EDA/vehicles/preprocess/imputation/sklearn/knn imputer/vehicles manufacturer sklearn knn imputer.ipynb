```python
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.model_selection import train_test_split
import numpy as np

vehicles_data_set = pd.read_csv('/home/jovyan/work/datasets/vehicles_cleaned.csv')
vehicles_data_set.head()
```

```python
X = vehicles_data_set.copy()

y = vehicles_data_set['price']
X.drop(columns=['price'], inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
cols_with_null_values = [col for col in X_train.columns
                       if X_train[col].isnull().any()]

for col_with_null_values in cols_with_null_values:
    print(
        X_train[col_with_null_values].value_counts(dropna=False)
    )
```

```python
X_train.reset_index(inplace=True, drop=True)
```

```python
rows_with_manufacturer_null_values = X_train[X_train['manufacturer'].isnull()].index.to_list()

X_train_manufacturer_knn_imputed = X_train.copy()

print(
    X_train_manufacturer_knn_imputed['manufacturer'].isnull().any()
)

knn_imputer = KNNImputer(add_indicator=True)

knn_imputer_values = knn_imputer.fit_transform(
    np.reshape(
        X_train_manufacturer_knn_imputed['manufacturer'],
        (-1,1)
    )
)

X_train_manufacturer_knn_imputed.drop(columns=['manufacturer'], inplace=True)

X_train_manufacturer_knn_imputed = pd.concat(
    [
        X_train_manufacturer_knn_imputed,
        pd.Series(
            knn_imputer_values[:,0],
            name='manufacturer'
        ),
        pd.Series(
            knn_imputer_values[:,1],
            name='manufacturer_missing'
        )
    ], axis=1
)

print(
    X_train_manufacturer_knn_imputed['manufacturer'].isnull().any()
)

for row_with_manufacturer_null_values in rows_with_manufacturer_null_values:
    print(
        X_train['manufacturer'][row_with_manufacturer_null_values],
        '------------------',
        X_train_manufacturer_knn_imputed['manufacturer'][row_with_manufacturer_null_values],
        '------------------',
        X_train_manufacturer_knn_imputed['manufacturer_missing'][row_with_manufacturer_null_values]
    )
```
