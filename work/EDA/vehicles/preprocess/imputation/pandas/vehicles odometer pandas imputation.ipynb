```python
import pandas as pd
from sklearn.model_selection import train_test_split

vehicles_data_set = pd.read_csv('/home/jovyan/work/datasets/vehicles_cleaned.csv')
vehicles_data_set
```

```python
y = vehicles_data_set['price']
X = vehicles_data_set.copy()
X.drop(columns=['price'], inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
X_train['odometer'].isnull().sum()
```

```python
index_rows_null = X_train[X_train['odometer'].isnull()].index.tolist()
X_train_odometer_imputed_ffill = X_train.copy()
X_train_odometer_imputed_bfill = X_train.copy()
X_train_odometer_imputed_interpolate_linear = X_train.copy()

print(
    X_train_odometer_imputed_ffill['odometer'].isnull().any(),
    X_train_odometer_imputed_bfill['odometer'].isnull().any(),
    X_train_odometer_imputed_interpolate_linear['odometer'].isnull().any()
)

X_train_odometer_imputed_ffill['odometer'].ffill(inplace=True)
X_train_odometer_imputed_bfill['odometer'].bfill(inplace=True)
X_train_odometer_imputed_interpolate_linear['odometer'].interpolate(inplace=True)

print(
    X_train_odometer_imputed_ffill['odometer'].isnull().any(),
    X_train_odometer_imputed_bfill['odometer'].isnull().any(),
    X_train_odometer_imputed_interpolate_linear['odometer'].isnull().any()
)

for index_row_null in index_rows_null:
    print(
        X_train_odometer_imputed_ffill['odometer'][index_row_null],
        '-----',
        X_train_odometer_imputed_bfill['odometer'][index_row_null],
        '-----',
        X_train_odometer_imputed_interpolate_linear['odometer'][index_row_null],
    )
```
