```python
import pandas as pd
from sklearn.model_selection import train_test_split

vehicles_data_set = pd.read_csv('/home/jovyan/work/datasets/vehicles_cleaned.csv')
vehicles_data_set
```

```python
X = vehicles_data_set.copy()
y = X['price']
X.drop(columns=['price'], inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
cols_with_null = [col for col in X_train.columns
                 if X_train[col].isnull().any()]

for col_with_null in cols_with_null:
    print(X_train[col_with_null].value_counts(dropna=False))
```

```python
type_rows_with_null_value = X_train[X_train['type'].isnull()].index.to_list() 

X_type_ffill_imputed = X_train.copy()
X_type_bfill_imputed = X_train.copy()
X_type_mode_imputed = X_train.copy()
X_type_other_imputed = X_train.copy()

print(
    X_type_ffill_imputed['type'].isnull().any(),
    X_type_bfill_imputed['type'].isnull().any(),
    X_type_mode_imputed['type'].isnull().any(),
    X_type_other_imputed['type'].isnull().any()
)

X_type_ffill_imputed['type'].ffill(inplace=True)
X_type_bfill_imputed['type'].bfill(inplace=True)
X_type_mode_imputed['type'].fillna(
    X_type_mode_imputed['type'].mode()[0],
    inplace=True
)
X_type_other_imputed['type'].fillna(
    'other',
    inplace=True
)

print(
    X_type_ffill_imputed['type'].isnull().any(),
    X_type_bfill_imputed['type'].isnull().any(),
    X_type_mode_imputed['type'].isnull().any(),
    X_type_other_imputed['type'].isnull().any()
)

for type_row_with_null_value in type_rows_with_null_value:
     print(
         X_type_ffill_imputed['type'][type_row_with_null_value],
         '-----------',
         X_type_bfill_imputed['type'][type_row_with_null_value],
         '-----------',
         X_type_mode_imputed['type'][type_row_with_null_value],
         '-----------',
         X_type_other_imputed['type'][type_row_with_null_value]
     )   
```
