```python
import pandas as pd
from sklearn.model_selection import train_test_split

vehicles_data_set = pd.read_csv('/home/jovyan/work/datasets/vehicles_cleaned.csv')
vehicles_data_set
```

```python
X = vehicles_data_set.copy()
y = vehicles_data_set['price']
X.drop(columns=['price'], inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
X_train['title_status'].unique()
```

```python
index_rows_null = X_train[X_train['title_status'].isnull()].index.to_list()

X_train_title_status_ffill_imputed = X_train.copy()
X_train_title_status_bfill_imputed = X_train.copy()
X_train_title_status_mode_imputed = X_train.copy()
X_train_title_status_missing_imputed = X_train.copy()

print(
    X_train_title_status_ffill_imputed['title_status'].isnull().any(),
    X_train_title_status_bfill_imputed['title_status'].isnull().any(),
    X_train_title_status_mode_imputed['title_status'].isnull().any(),
    X_train_title_status_missing_imputed['title_status'].isnull().any()
)

X_train_title_status_ffill_imputed['title_status'].ffill(inplace=True)
X_train_title_status_bfill_imputed['title_status'].bfill(inplace=True)
X_train_title_status_mode_imputed['title_status'].fillna(
    X_train_title_status_mode_imputed['title_status'].mode()[0], 
    inplace=True
)
X_train_title_status_missing_imputed['title_status'].fillna(
    'missing', 
    inplace=True
)

print(
    X_train_title_status_ffill_imputed['title_status'].isnull().any(),
    X_train_title_status_bfill_imputed['title_status'].isnull().any(),
    X_train_title_status_mode_imputed['title_status'].isnull().any(),
    X_train_title_status_missing_imputed['title_status'].isnull().any()
)

for index_row_null in index_rows_null:
    print(
        X_train_title_status_ffill_imputed['title_status'][index_row_null],
        '---------',
        X_train_title_status_bfill_imputed['title_status'][index_row_null],
        '---------',
        X_train_title_status_mode_imputed['title_status'][index_row_null],
        '---------',
        X_train_title_status_missing_imputed['title_status'][index_row_null], 
    )   
```
