```python
import pandas as pd
from sklearn.model_selection import train_test_split

vehicles_data_set = pd.read_csv('/home/jovyan/work/datasets/vehicles_cleaned.csv')
vehicles_data_set
```

```python
X = vehicles_data_set.copy()
y = X['price']
X.drop(columns=['price'], inplace=True)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
cols_with_null_values = [col for col in X_train.columns
                       if X_train[col].isnull().any()]
for col_with_null_values in cols_with_null_values:
    print(X_train[col_with_null_values].value_counts(dropna=False))
```

```python
indexes_row_fuel_null_values = X_train[X_train['fuel'].isnull()].index.to_list()

X_train_ffill_imputed = X_train.copy()
X_train_bfill_imputed = X_train.copy()
X_train_mode_imputed = X_train.copy()
X_train_other_imputed = X_train.copy()

print(
    X_train_ffill_imputed['fuel'].isnull().sum(),
    X_train_bfill_imputed['fuel'].isnull().sum(),
    X_train_mode_imputed['fuel'].isnull().sum(),
    X_train_other_imputed['fuel'].isnull().sum()
)

X_train_ffill_imputed['fuel'].ffill(inplace=True)
X_train_bfill_imputed['fuel'].bfill(inplace=True)
X_train_mode_imputed['fuel'].fillna(
    X_train_mode_imputed['fuel'].mode()[0],
    inplace=True
)
X_train_other_imputed['fuel'].fillna(
    'other',
    inplace=True
)

print(
    X_train_ffill_imputed['fuel'].isnull().sum(),
    X_train_bfill_imputed['fuel'].isnull().sum(),
    X_train_mode_imputed['fuel'].isnull().sum(),
    X_train_other_imputed['fuel'].isnull().sum()
)

for index_row_fuel_null_values in indexes_row_fuel_null_values:
    print(
        X_train_ffill_imputed['fuel'][index_row_fuel_null_values],
        '----------------',
        X_train_bfill_imputed['fuel'][index_row_fuel_null_values],
        '----------------',
        X_train_mode_imputed['fuel'][index_row_fuel_null_values],
        '----------------',
        X_train_other_imputed['fuel'][index_row_fuel_null_values]
    )
```
