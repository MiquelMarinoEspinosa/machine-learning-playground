```python
import pandas as pd
from sklearn.model_selection import train_test_split

vehicles_data_set = pd.read_csv('/home/jovyan/work/datasets/vehicles.csv')
vehicles_data_set.head()
```

```python
y = vehicles_data_set['price']
X = vehicles_data_set.copy()
X.drop(columns=[
    'price',
    'region_url', 
    'description', 
    'image_url', 
    'url',
    'lat',
    'long',
    'VIN',
    'posting_date'
], inplace=True, errors='ignore')

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.6,
    random_state=42
)

X_test, X_val, y_test, y_val = train_test_split(
    X_test,
    y_test,
    train_size=0.5,
    random_state=42
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape, X_val.shape, y_val.shape
```

```python
cols_with_null = [col for col in X_train.columns
                if X_train[col].isnull().any()]

for col_with_null in cols_with_null:
    print(X_train[col_with_null].value_counts(dropna=False))
    print(X_train[col_with_null].isnull().sum())
```
